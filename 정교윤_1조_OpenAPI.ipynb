```
# API Test
# open_api_url = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcNrgTrade?LAWD_CD=11110&DEAL_YMD=201512&serviceKey=dHqjQmlYRwI0M12xB1y2qIYpliyjgyq5oK0FvzeVHbpTi665iJxzGO%2B3ZL6iLY60xeOFuD2TfUo0RZVdc%2B9Mfg%3D%3D'

# 법정동코드 
# LAWD_CD = '11110'

# 계약년월일
# DEAL_YMD = '201512'

# 인증키 
serviceKey = 'dHqjQmlYRwI0M12xB1y2qIYpliyjgyq5oK0FvzeVHbpTi665iJxzGO%2B3ZL6iLY60xeOFuD2TfUo0RZVdc%2B9Mfg%3D%3D'

def get_search_result_data_frame(LAWD_CD, DEAL_YMD, serviceKey):
  
  open_api_url = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcNrgTrade?LAWD_CD='+LAWD_CD+'&DEAL_YMD='+DEAL_YMD+'&serviceKey='+serviceKey


#open_api_url = 'http://openapi.molit.go.kr/OpenAPI_ToolInstallPackage/service/rest/RTMSOBJSvc/getRTMSDataSvcNrgTrade'


#call_url = open_api_url + '?'

# 파라미터 연결, utf-8 선언안하면 ascii 코드로 변환되어서 키값이 변조될 수 있습니다.
#params = urlencode({quote_plus('LAWD_CD'):LAWD_CD, quote_plus('DEAL_YMD') : DEAL_YMD, quote_plus('serviceKey') : serviceKey}, encoding='utf-8')

# urllib의 Request로 요청 url로 병합
#request = Request(call_url + params)

  request = Request(open_api_url)

  request.get_method = lambda: 'GET'
  response_body = urlopen(request).read().decode('utf-8')

  parsed = xmltodict.parse(response_body, dict_constructor=dict)
  
  df = pd.DataFrame.from_dict(parsed['response']['body']['items']['item'], orient='columns')
  
  # csv로 저장 
  df.to_csv('search_result.csv',sep=',')
  
  return df
```
